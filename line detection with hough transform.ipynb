```python
x = range(20)
m_tag = 3
b = 5
```

```python
y = [m_tag*coeff + b for coeff in x]
y
```

```python
def get_noise_with_x(x, random_coeff):
    return [i + random.randint(-random_coeff, random_coeff) for i in x] 
```

```python
def plot_x_y_graph(x, y, random_coeff, without_noise=False):
    x = get_noise_with_x(x, random_coeff)
    if without_noise:
        plt.plot(x, y, linestyle='--', marker='o')
    else:
        plt.plot(x, y, linestyle='--')
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.show()
```

```python
def plot_m_b_graph(x, y, m):
    plt.xlabel("m")
    plt.ylabel("b")
    for _x, _y in zip(x, y):
        b_star = [_y - m_index*_x for m_index in m]
        print(b_star)
        plt.plot(m, b_star, linestyle='--')
    plt.show()
    plt.savefig("m_b")
```

### basic x-y

```python
plot_x_y_graph(x, y, 0, without_noise=True)
```

### b-m representation

```python
m = range(-4, 10)
plot_m_b_graph(x, y, m)
```

### x-y representation with noise

```python
plot_x_y_graph(x, y, 2)
```

### m-b representation with noise

```python
for rand in range(1, 10, 3):
    x_noise = get_noise_with_x(x, rand)
    plot_m_b_graph(x_noise, y, m)
```

### theta representation with shape of (1280, 800)

```python
thetas = np.arange(-90, 90)
image_size_X = 1280
image_size_Y = 800
x_theta_trans = range(0, image_size_X, 32)
y_theta_trans = range(0, image_size_Y, 20)
```

```python
plt.xlabel("theta")
plt.ylabel("ro")
for _x, _y in zip(x_theta_trans, y_theta_trans):
    ro = [_x*cos(radians(theta)) + _y*sin(radians(theta)) for theta in thetas]
    img = plt.plot(thetas, ro, linestyle='--')
plt.show()
plt.savefig("ro_theta")
```

The benefit is that we don't have to search the limits of the value of 'm'.

```python
rho_max = int(np.ceil(np.sqrt(image_size_X**2 + image_size_Y**2))) 
delta_rho = 20  # rho bin pixels
delta_theta = 10  # theta bin size in degrees

rho_bins = int(np.ceil(2 * rho_max / delta_rho))
theta_bins = int(np.ceil(180 / delta_theta))

hough_hist_2D = np.zeros((rho_bins, theta_bins))

theta_values = np.arange(-90 + delta_theta / 2, 90 - delta_theta / 2, delta_theta)

for index in range(len(x_theta_trans)):
    _x = x_theta_trans[index]
    _y = y_theta_trans[index]
    rho = _x* np.cos(np.deg2rad(theta_values)) + _y* np.sin(np.deg2rad(theta_values))
    rho_bin_number = np.around((rho + rho_max) / delta_rho)
    for theta_indx in range(len(theta_values)):
        hough_hist_2D[int(rho_bin_number[theta_indx]), theta_indx] += 1
hough_hist_2D
```

```python
plt.figure(7)
plt.plot(hough_hist_2D)
plt.show()
```
